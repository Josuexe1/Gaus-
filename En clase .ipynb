```python
import matplotlib.pyplot as plt

%autoreload 2

# Datos del sistema
A = ([[3, 2], [2, 10]])
b = ([11, 16])
x0 =([1, 3])
tol = 1e-1
max_iterations = 100


# Llamada a la función gauss_jacobi
x_jacobi, tray_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iterations)

# Llamada a la función gauss_seidel
x_seidel, tray_seidel = gauss_seidel(A=A, b=b, x0=x0, tol=tol, max_iter=max_iterations)

# Mostrar resultados
print("Solución aproximada Gauss-Jacobi:")
print(x_jacobi)
print("Trayectoria de iteraciones Gauss-Jacobi:")
for i, xi in enumerate(tray_jacobi):
    print(f"Iteración {i}: {xi.T}")

print("\nSolución aproximada Gauss-Seidel:")
print(x_seidel)
print("Trayectoria de iteraciones Gauss-Seidel:")
for i, xi in enumerate(tray_seidel):
    print(f"Iteración {i}: {xi.T}")

# Graficar las trayectorias de iteraciones para x[0] de ambos métodos
iterations_jacobi = list(range(len(tray_jacobi)))
values_x0_jacobi = [xi[0][0] for xi in tray_jacobi]  # Trayectoria de x[0] Gauss-Jacobi
values_x1_jacobi = [xi[1][0] for xi in tray_jacobi]  # Trayectoria de x[1] Gauss-Jacobi

iterations_seidel = list(range(len(tray_seidel)))
values_x0_seidel = [xi[0][0] for xi in tray_seidel]  # Trayectoria de x[0] Gauss-Seidel
values_x1_seidel = [xi[1][0] for xi in tray_seidel]  # Trayectoria de x[1] Gauss-Seidel

plt.figure(figsize=(8, 6))

# Graficar la trayectoria de iteraciones para Gauss-Jacobi
plt.plot(values_x0_jacobi, values_x1_jacobi, linestyle='--', marker='o', label='Gauss-Jacobi')

# Graficar la trayectoria de iteraciones para Gauss-Seidel
plt.plot(values_x0_seidel, values_x1_seidel, linestyle='-', marker='x', label='Gauss-Seidel')

# Resaltar el punto inicial para Gauss-Jacobi
plt.scatter(values_x0_jacobi[0], values_x1_jacobi[0], color='blue', marker='s', s=150, label='Punto inicial Gauss-Jacobi', edgecolors='black', linewidths=2)

# Resaltar el punto final para Gauss-Jacobi
plt.scatter(values_x0_jacobi[-1], values_x1_jacobi[-1], color='red', marker='o', s=150, label='Solución final Gauss-Jacobi', edgecolors='black', linewidths=2)

# Resaltar el punto inicial para Gauss-Seidel
plt.scatter(values_x0_seidel[0], values_x1_seidel[0], color='green', marker='s', s=150, label='Punto inicial Gauss-Seidel', edgecolors='black', linewidths=2)

# Resaltar el punto final para Gauss-Seidel
plt.scatter(values_x0_seidel[-1], values_x1_seidel[-1], color='orange', marker='o', s=150, label='Solución final Gauss-Seidel', edgecolors='black', linewidths=2)

# Configurar etiquetas y título
plt.xlabel('x[0]')
plt.ylabel('x[1]')
plt.title('Trayectoria de iteraciones de Gauss-Jacobi y Gauss-Seidel')
plt.grid(True)
plt.legend()
plt.show()
```
